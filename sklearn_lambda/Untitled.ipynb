```python
import pandas as pd

import matplotlib.pyplot as plt
 
plt.rcParams['figure.figsize']=(20,10)
plt.style.use('ggplot')


%matplotlib inline  
```

```python
df = pd.read_csv('../examples/climate-change-earth-surface-temperature-data/GlobalTemperatures.csv', index_col='dt', parse_dates=True)
```

```python
df['landavgTemp_MA'] = df.LandAverageTemperature.rolling(window=24, center=False).mean()
```

```python
df.tail()
```

```python
df[['LandAverageTemperature', 'landavgTemp_MA']].plot(figsize=(20,10))
```
